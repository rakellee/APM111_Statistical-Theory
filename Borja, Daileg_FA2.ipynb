```python
import numpy as np

X = np.array([2, 4, 6, 8, 10])
N = len(X)

lhs = np.sum((X - 1) ** 2)
rhs = np.sum(X ** 2) - 2 * np.sum(X) + N

print("Exercises")
print("-" * 40)
print("Problem: 3.49")
print("-" * 40)
print("Verification of Identity: \nΣ(X - 1)² = ΣX² - 2ΣX + N \n")
print(f"LHS = {lhs}")
print(f"RHS = {rhs}")
print("Identity holds:", lhs == rhs)

print()
print()

U = np.array([3, -2, 5])
V = np.array([-4, -1, 6])

a = np.sum(U * V)
b = np.sum((U + 3) * (V - 4))
c = np.sum(V ** 2)
d = ((np.sum(U)) * (np.sum(V)) ** 2)
e = np.sum(U * (V ** 2))
f = np.sum((U ** 2) - 2 * (V **2) + 2)
g = np.sum(U / V)

print("-" * 40)
print("Problem: 3.51")
print("-" * 40)
print(f"(a) ΣUV             = {a}")
print(f"(b) Σ(U+3)(V−4)     = {b}")
print(f"(c) (ΣU)(ΣV)²       = {c}")
print(f"(d) (ΣU)²           = {d}")
print(f"(e) ΣUV²            = {e}")
print(f"(f) Σ(U² - 2V² + 2) = {f}")
print(f"(g) Σ(U/V)          = {g}")

print()
print()

set_a = np.array([3, 5, 8, 3, 7, 2])
set_b = np.array([28.5, 73.6, 47.2, 31.5, 64.8])

gm_a = np.prod(set_a) ** (1 / len(set_a))
gm_b = np.prod(set_b) ** (1 / len(set_b))

print("-" * 40)
print("Problem: 3.90")
print("-" * 40)
print(f"Geometric Mean (Set a): {gm_a:.4f}")
print(f"Geometric Mean (Set b): {gm_b:.4f}")
```
